```python
combined_csv =pd.read_csv('../data/combined.csv')
```

```python
combined_csv.describe()
```

```python
for i in combined_csv.columns: 
    print(i,':', combined_csv[i].isnull().any())
```

```python
combined_csv["ended_at"] = pd.to_datetime(combined_csv["ended_at"], unit='ns')
combined_csv["started_at"] = pd.to_datetime(combined_csv["started_at"], unit='ns')
combined_csv.info()
```

```python
combined_csv["trip_duration"] = (combined_csv["ended_at"] - combined_csv["started_at"])/pd.Timedelta(minutes=1)
combined_csv["day_of_week"] = combined_csv["started_at"].dt.day_name()
combined_csv["trip_month"] = combined_csv["started_at"].dt.month_name()
combined_csv.head()
```

```python
df = combined_csv[combined_csv['trip_duration'] >= 1]
```

```python
df = df.drop(['end_station_name','start_station_name','end_station_id','start_station_id','end_lat', 'end_lng'], axis = 1)
```

```python
df.to_csv('../data/min_rides.csv')
```

```python
combined_csv.loc[(combined_csv['trip_duration'] >= 1)].groupby( ['member_casual'], as_index=False ) \
    .agg({'trip_duration':'mean','ride_id':'count'}) \
    .sort_values('ride_id', ascending=False) \
```

```python
combined_csv.loc[(combined_csv['trip_duration'] >= 1)].groupby( ['member_casual', 'day_of_week'], as_index=False ) \
    .agg({'trip_duration':'mean','ride_id':'count'}) \
    .sort_values('ride_id', ascending=False).set_index(['day_of_week', 'member_casual']) \
```

```python
combined_csv.loc[(combined_csv['trip_duration'] >= 1)].groupby( ['member_casual', 'trip_month'], as_index=False ) \
    .agg({'trip_duration':'mean','ride_id':'count'}) \
    .sort_values('ride_id', ascending=False).set_index(['trip_month', 'member_casual'])  \
```

```python
combined_csv['daytime'] = pd.qcut(combined_csv['started_at'].dt.hour, 4, labels=["morning", "afternoon", "evening", "night"])
```

```python
combined_csv.loc[(combined_csv['trip_duration'] >= 1)].groupby( ['member_casual', 'daytime'], as_index=False ) \
    .agg({'trip_duration':'mean','ride_id':'count'}) \
    .sort_values('ride_id', ascending=False).set_index(['daytime', 'member_casual'])  \
```
